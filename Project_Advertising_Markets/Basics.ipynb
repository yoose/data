## Evaluating Marketing Opportunities for an E-learning Company

We are an e-learning company that offers courses on programming. The company would like to spend some money on advertising and would like to know how to spend the money. The goal of this project is to determine what the two best markets to advertise our products in are.

## Understanding the data

Obtaining our own data is very expensive so for the purpose of this study we will looking at existing data to see if it can serve our purposes.

The data that we will use is a survey that freeCodeCamp conducted, the [2017 New Coder Survey](https://medium.freecodecamp.org/we-asked-20-000-people-who-they-are-and-how-theyre-learning-to-code-fff5d668969). The survey data is publicly available at [this GitHub repository](https://github.com/freeCodeCamp/2017-new-coder-survey). Below we will take a look at the survey data from the 'clean-data' folder at the previously mentioned repository. The data can be downloaded directly at this [link](https://raw.githubusercontent.com/freeCodeCamp/2017-new-coder-survey/master/clean-data/2017-fCC-New-Coders-Survey-Data.csv).

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

file_link = direct_link = 'https://raw.githubusercontent.com/freeCodeCamp/2017-new-coder-survey/master/clean-data/2017-fCC-New-Coders-Survey-Data.csv'
data = pd.read_csv(file_link, low_memory = 0)
print(data.shape)
pd.options.display.max_columns = 150 # to avoid truncated output 
data.head()
```

Only about 31.7% of the survey participants are interested in only one subject. These numbers suggest that this data contains a sample that is representative of our population of interest. We want to advertise to people who want are interested in many different forms of programming, but mainly web and mobile development.

Now we want to investigate further to determine which are the two best markets to advertise in. To do this, we might look at the following:

* Where are these aspiring developers located?
* Which locations contain the highest number of potential new customers?
* How much are people willing to spend on coding education?
* Which age groups contain the most potential customers?


# Potential Customers - Locations and Densities

To begin our analysis, let's take a look at where they are located and the densities of those locations. This should be a good first step in finding the two best markets. We will start by looking at the CountryLive column to determine where the participants are located. Also, we will only be considering participants who provided a reply for which role(s) they are interested in.

```python
# Maka a copy of the dataframe containing only participants that replied with role(s) of interest
data_replied = data[data['JobRoleInterest'].notnull()].copy()

# Frequency tables of the countries
abs_freq = data_replied['CountryLive'].value_counts()
rel_freq = data_replied['CountryLive'].value_counts(normalize=True) * 100

# Display the frequency tables
pd.DataFrame(data = {'Absolute Frequency': abs_freq,
                    'Percentage': rel_freq})
```

The US contains 45.7% of our potential customers. Coming in a far second is India with 7.72%, but not too far behind are the UK and Canada with 4.6% and 3.8% respectively. All the other countries are under 2%. While it is clear that the US is the biggest market the next three countries are close enough that we will need to look at additional data to determine the second market we want to advertise in. While India has the second highest percentage of potential customers, we also need to consider whether or not and how much customers are willing to pay for courses. It would be a waste of money to advertise in a market where the majority of potential customers are not willing to pay for courses.

# Spending money on courses

We are planning to sell subscriptions to our online learning platform for $59 per month, so if possible we would like to determine how much students have spent so far on courses. In addition, since our courses will be taught in English and the four countries with the highest absolute frequencies have English as one of their official languages, we will be focusing our analysis on the US, India, the UK, and Canada.

We will be looking at the MoneyForLearning and MonthsProgramming columns for this analysis. The MoneyForLearning column contains, in US dollars, the amount of money spent by participants from when they started coding until they took the survey. The MonthsProgramming contains the number of months they have been coding. Taking a quick look at the MonthsProgramming column we can see that some participants replied 0; to avoid dividing by 0 we will be replacing the 0s with 1s. It is possible that the participants replied 0 because it had not been a full month when they filled out the survey.

```python
data_replied['MonthsProgramming'].value_counts().sort_index()
```

```python
# Replace 0s with 1s to avoid dividing by 0
data_replied['MonthsProgramming'].replace(0,1, inplace=True)

# Create a new column for money spent per month
data_replied['amt_per_month'] = data_replied['MoneyForLearning'] / data_replied['MonthsProgramming']
print(data_replied['amt_per_month'].isnull().sum())

# Remove participants that did not spend any money
data_replied = data_replied[data_replied['amt_per_month'].notnull()]

# Remove participants that did not provide a reply to the CountryLive column
data_replied = data_replied[data_replied['CountryLive'].notnull()]

# Sanity check to see how much data we have left
data_replied['CountryLive'].value_counts().head(4)
```

```python
# Lets take a look mean amount of money spend per month, grouped by countries
countries_mean = data_replied.groupby('CountryLive').mean()
countries_mean['amt_per_month'][['United States of America',
                            'India', 'United Kingdom',
                            'Canada']]
```

```python
usa_median = data_replied[data_replied['CountryLive'] == 'United States of America']['amt_per_month'].median()
india_median = data_replied[data_replied['CountryLive'] == 'India']['amt_per_month'].median()
uk_median = data_replied[data_replied['CountryLive'] == 'United Kingdom']['amt_per_month'].median()
canada_median = data_replied[data_replied['CountryLive'] == 'Canada']['amt_per_month'].median()

print('US', usa_median)
print('India', india_median)
print('UK', uk_median)
print('Canada', canada_median)
#print(data_replied[data_replied['CountryLive'] == 'India']['amt_per_month'])
```

```python
usa_mode = data_replied[data_replied['CountryLive'] == 'United States of America']['amt_per_month'].mode()
india_mode = data_replied[data_replied['CountryLive'] == 'India']['amt_per_month'].mode()
uk_mode = data_replied[data_replied['CountryLive'] == 'United Kingdom']['amt_per_month'].mode()
canada_mode = data_replied[data_replied['CountryLive'] == 'Canada']['amt_per_month'].mode()

print('US', usa_mode)
print('India', india_mode)
print('UK', uk_mode)
print('Canada', canada_mode)
```

```python
usa_count = data_replied[data_replied['CountryLive'] == 'United States of America']['amt_per_month'].value_counts().sort_index()
india_count = data_replied[data_replied['CountryLive'] == 'India']['amt_per_month'].value_counts().sort_index()
uk_count = data_replied[data_replied['CountryLive'] == 'United Kingdom']['amt_per_month'].value_counts().sort_index()
canada_count = data_replied[data_replied['CountryLive'] == 'Canada']['amt_per_month'].value_counts().sort_index()

print('US', '\t', usa_count[0], '\t', 2933, '\t', usa_count[0]/2933 * 100)
print('India', '\t', india_count[0], '\t', 463, '\t', india_count[0]/463 * 100)
print('UK', '\t', uk_count[0], '\t', 279, '\t', uk_count[0]/279 * 100)
print('Canada', '\t', canada_count[0], '\t', 240, '\t', canada_count[0]/240 * 100)
```

It would appear that in addition to having the biggest markets, the US and India also spend the most per month for coding courses. While it is not surprising for the US to be at the top, it is a bit unanticipated that India exceeds the UK and Canada in spending given that the UK and Canada have a much higher GDP than India, which is considered a third world country. We should take a deeper look at the data to see if we can explain this. One thing that we can look at is to see if there are any outliers in our data.

Another interesting note is that each country has at least 45% of participants that have not spent any money on learning with India having the most at 56.6% and the US with the least at 44.8%. While we can not determine whether or not those participants are willing to spend money, we can do an analysis on only those that spent money to compare.

```python
# Isolate only the countries of interest
top4 = data_replied[data_replied['CountryLive'].str.contains(
    'United States of America|India|United Kingdom|Canada')]

# We will use box plots to visualize our data
import seaborn as sns
sns.boxplot(y = 'amt_per_month', x = 'CountryLive',
            data = top4)
plt.title('Amount Spent Per Month By Country\n(Distributions)',
         fontsize = 16)
plt.ylabel('Amount per month (US dollars)')
plt.xlabel('Country')
plt.xticks(range(4), ['US', 'UK', 'India', 'Canada']) # avoids tick labels overlap
plt.show()
```

We created a box plot to visualize the distribution of the data, but from our first plot, we can not ascertain much about the India, the UK, and Canada. What we can see is that there are some outliers in our US data; there are two participants that spent over \$50,000 per month in education. Taking a quick look at those two participants, they have only been studying programming for 1 or 2 months and have spent a lot of money, perhaps to join a programmer, which is why the amt_per_month is so high. One spent \$100k and the other \$80k which is a lot for anything short of a university degree program and the question specifies that it should be money spent aside from Universtiy Tuition. I think it is safe to treat these data points as outliers and remove them from our analysis.

```python
data_replied = data_replied[data_replied['amt_per_month'] < 20000]

# Lets take a look at the updated means
countries_mean = data_replied.groupby('CountryLive').mean()
countries_mean['amt_per_month'][['United States of America',
                            'India', 'United Kingdom',
                            'Canada']]
```

```python
# Updated Box Plots
# Isolate only the countries of interest
top4 = data_replied[data_replied['CountryLive'].str.contains(
    'United States of America|India|United Kingdom|Canada')]

# We will use box plots to visualize our data
sns.boxplot(y = 'amt_per_month', x = 'CountryLive',
            data = top4)
plt.title('Amount Spent Per Month By Country\n(Distributions)',
         fontsize = 16)
plt.ylabel('Amount per month (US dollars)')
plt.xlabel('Country')
plt.xticks(range(4), ['US', 'UK', 'India', 'Canada']) # avoids tick labels overlap
plt.show()
```

It looks like there are a few outliers for India, perhaps above the $3,000 range and one for Canada. Let's take a look at the possible outliers for India and see if we can determine the reasons for the high spending, such as attending a bootcamp.

```python
# Inspect the outliers for India
india_outliers = top4[
    (top4['CountryLive'] == 'India') & 
    (top4['amt_per_month'] >= 3000)]
india_outliers
```

None of the outliers said they have attended a bootcamp. Four of the students spent \$20,000 or more on education, which is a significant amount of money to spend for someone living in India. One possible explanation are online degrees, which participants might have thought is different from University tuition. The other two numbers are exceptionally high because they have only been programming for 1 month, so the average per month is high. I think it is safe to remove these outliers from our sample.

```python
# Remove the outliers for India
top4 = top4.drop(india_outliers.index) # using the row labels
```

Looking at the updated box plot, we can see there are still some extreme outliers for the US at about \$6,000 and above so let's take a closer look at those.

```python
# Examine the extreme outliers for the US
us_outliers = top4[
    (top4['CountryLive'] == 'United States of America') & 
    (top4['amt_per_month'] >= 6000)]

us_outliers
```

Of the 11 outliers above, six replied that they attended bootcamps and all six of those appear to have started recently as they all have three months or less of coding experience. Since bootcamp tuitions are probably paid up front, the amount spent per month are extremely high and can be removed.

For those that did not attend a bootcamp the amount spent are extremely high, \$30,000+ or they have only been studying for a short amount of time, resulting in a high amount spent per month like those mentioned earlier.

So it should be safe to remove all of these outliers.

```python
# Remove extreme outliers from the US
top4 = top4.drop(us_outliers.index) # using the row labels
```

Taking a look again at the updated box plot, there appears to be one extreme outlier so we will examine that point.

```python
# Examine the extreme outliers for Canada
canada_outliers = top4[
    (top4['CountryLive'] == 'Canada') & 
    (top4['amt_per_month'] > 4500)]

canada_outliers
```

This outlier is similar to some of the data points that we saw earlier in the US data, they spent a large sum on a bootcamp, but have only recently started so we will remove this outlier.

```python
# Remove extreme outliers from Canada
top4 = top4.drop(canada_outliers.index) # using the row labels
```

```python
# Recompute mean sum of money spent by students each month
top4.groupby('CountryLive').mean()['amt_per_month']
```

# Selecting the two best markets to advertise our e-learning platform

The obvious first choice here is the US. There are a lot of potential customers and they are willing to spend the most per month, ~\$143 per month. Looking purely at the amount spent per month, Canada, at \$93 per month, would be the second choice to advertise in. However, we should take into consideration that India has a larger population and more potential customers. Their average amount spent is ~\$65 per month, which is above our target subscription price.

Possible courses of action:
1. Advertise in the US, Canada, and India by splitting the advertising budget among all three markets. Some examples of splitting the budget are:
    * US: 60%, Canada: 30%, India: 10%
    * US: 60%, Canada: 10%, India: 30%
    * US: 70%, Canada: 15%, India: 15%

2. Advertise in the US and either Canada or India. The following should be taken into consideration to determine whether Canada or India should be the second market.
    * On average Canada spends more per month to learn coding than India.
    * India has a much larger market than Canada.
    * India has a higher percentage of people than Canada who have not spent any money on learning, 56.5% to 52.5%.
